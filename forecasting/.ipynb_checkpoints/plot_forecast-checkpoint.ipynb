```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import to_date, year, month, col

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Start Spark
spark = (
    SparkSession.builder
    .appName("PlotForecastNotebook")
    .getOrCreate()
)
```

```python
# 2) Load street data from HDFS & aggregate to monthly counts
df = (
    spark
      .read
      .parquet("hdfs://localhost:9000/user/yashwanthreddy/crime/clean/street")
      .withColumn("date", to_date(col("date")))
)

monthly = (
    df
      .groupBy(year("date").alias("year"), month("date").alias("month"))
      .count()
      .orderBy("year", "month")
      .toPandas()
)

# Build a proper datetime index
monthly["date"] = pd.to_datetime(dict(
    year=monthly["year"],
    month=monthly["month"],
    day=1
))
monthly.set_index("date", inplace=True)
ts = monthly["count"].sort_index()
```

```python
# 3) Tear down Spark
spark.stop()
```

```python
# 4) Load ARIMA forecast (mean + 95% CI)
fc = pd.read_csv("output/forecast.csv", index_col=0, parse_dates=True)
```

```python
# 5) Plot with improvements
plt.figure(figsize=(12, 6))

# Historical (blue)
plt.plot(ts.index, ts.values, label="Historical", color="C0")

# Forecast mean (orange)
plt.plot(fc.index, fc["mean"], label="Forecast", color="C1")

# Semi-transparent CI band (25% alpha)
plt.fill_between(
    fc.index,
    fc["mean_ci_lower"],
    fc["mean_ci_upper"],
    color="C1",
    alpha=0.25,
    label="95% CI"
)

# Titles, labels, grid, legend
plt.title("Monthly Crime Counts: Historical and 6-Month Forecast")
plt.xlabel("Date")
plt.ylabel("Number of Incidents")
plt.grid(True)
plt.legend(loc="upper left")

# Quarterly x-ticks
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.MonthLocator(bymonth=[1,4,7,10]))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```
